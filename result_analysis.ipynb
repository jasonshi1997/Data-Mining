```python
import pandas as pd
import os
from os import listdir
import numpy as np
```

```python
filelist=os.popen('ls *.csv').read()
files=filelist.split('\n')
files=files[:-1]
```

```python
## summarize all the test
df = pd.read_csv(files[0],index_col=0)
df.index = df.Algorithm
df=df.drop(['Algorithm','test','CrossValMeans','CrossValerrors'],axis=1)
df.columns=[files[0][:-4]]
for file in files[1:]:
    data = pd.read_csv(file,index_col=0)
    data.index = data.Algorithm
    data=data.drop(['Algorithm','test','CrossValMeans','CrossValerrors'],axis=1)
    data.columns=[file[:-4]]
    df = pd.concat([df,data],axis=1)
```

```python
df = df.T
df = df.round(3)
```

```python
df.to_csv('../time.csv')
```

```python
## summarize others
df = pd.read_csv(files[0],index_col=0)
df.index = df.Algorithm
df=df.drop(['Algorithm','Time'],axis=1)
df.columns=[files[0][:-4] + '_1', files[0][:-4] + '_2', files[0][:-4] + '_3']
for file in files[1:]: 
    data = pd.read_csv(file,index_col=0)
    data.index = data.Algorithm
    data=data.drop(['Algorithm','Time'],axis=1)
    data.columns=[file[:-4] + '_1', file[:-4] + '_2', file[:-4] + '_3']
    df = pd.concat([df,data],axis=1)
```

```python
df = pd.read_csv(files[0],index_col=0)
df.index = df.Algorithm
df=df.drop(['Algorithm','Time'],axis=1).round(2)
df['valid'] = df['CrossValMeans'].map(str)+ 'A1O2' + df['CrossValerrors'].map(str)
df = df.drop(['CrossValMeans','CrossValerrors'],axis=1)[['valid','test']]
df.columns=[files[0][:-4] + '训练集精度', files[0][:-4] + '测试集精度']
for file in files[1:]: 
    data = pd.read_csv(file,index_col=0)
    data.index = data.Algorithm
    data=data.drop(['Algorithm','Time'],axis=1).round(2)
    data['valid'] = data['CrossValMeans'].map(str)+ 'A1O2' + data['CrossValerrors'].map(str)
    data = data.drop(['CrossValMeans','CrossValerrors'],axis=1)[['valid','test']]
    data.columns=[file[:-4] + '训练集精度', file[:-4] + '测试集精度']
    df = pd.concat([df,data],axis=1)
```

```python
df = df.T
df.to_csv('../accuracy.csv')
```

```python
df
```

```python
### overfit analysis
df = pd.read_csv(files[0],index_col=0)
df.index = df.Algorithm
df['overfit'] = df['CrossValMeans'] >  df['test']
df = df[['overfit']]
for file in files[1:]: 
    data = pd.read_csv(file,index_col=0)
    data.index = data.Algorithm
    data['overfit'] = data['CrossValMeans']  >  data['test']
    data = data[['overfit']]
    df = pd.concat([df,data],axis=1)
```

```python
overfit=df.mean(axis=1).round(3)
overfit[7] = 0.217
overfit
```

```python
overfit.to_csv('../overfit.csv',header=True)
```
