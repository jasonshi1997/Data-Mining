```python
import pandas as pd
import os
from os import listdir
import numpy as np
```

```python
listdir()
```

```python
filelist=os.popen('ls *.csv').read()
files=filelist.split('\n')
files=files[:-1]
combine = pd.read_csv(files[0],index_col=0)
for file in files:
    df = pd.read_csv(file,index_col=0)
    combine = pd.concat([combine,df],axis=1)
combine
```

```python
combine=combine.iloc[:,1:]
```

```python
combine = combine.T
combine['dimension'] = combine['features'] / (combine['train_number'] + combine['test_number'])
combine['class_size'] = (combine['train_number'] + combine['test_number']) / combine['classes']
```

```python
combine.to_csv('../summary.csv')
```

```python
combine['number'] = combine['train_number'] + combine['test_number']
combine
```

```python
#combine=combine.drop(['train_number','test_number','outliers'])
combine = combine[['number','features','classes','class_size','dimension','binary','categorical','continous','meancor','classentropy']]
```

```python
decimals = pd.Series([0,0,0,1,4,2,2,2,2,2], index=['number','features','classes','class_size','dimension','binary','categorical','continous','meancor','classentropy'])
combine = combine.round(decimals)
```
